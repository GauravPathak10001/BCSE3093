```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
```

```python
y = 2 *x
y1 = 3 * x
```

```python
plt.plot(x,y,color ='green',linestyle=':',linewidth = 5)
plt.plot(x,y1,color = 'blue', linestyle = '-',linewidth=3)
plt.title('line plot')
plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.show()
```
